```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader

model_type = "MiDaS_small"
model = torch.hub.load("intel-isl/MiDaS", model_type)

device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
model.to(device)
print()
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import cv2

def save_depth(filename, save_path):
    img = cv2.imread(filename)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    transform = torch.hub.load("intel-isl/MiDaS", "transforms").small_transform
    batch = transform(img).to(device)

    with torch.no_grad():
        prediction = model(batch)

        prediction = torch.nn.functional.interpolate(
            prediction.unsqueeze(1),
            size=img.shape[:2],
            mode="bicubic",
            align_corners=False,
        ).squeeze()

    output = prediction.cpu().numpy()

    # np.save(save_path, output)

    # Create a figure without axes
    fig, ax = plt.subplots()
    ax.axis('off')  # Turn off the axes

    # Display the array as an image
    img = ax.imshow(output, cmap='gray_r', interpolation='nearest')
    plt.savefig(save_path, bbox_inches='tight', pad_inches=0)
    # plt.show()
    
    # fig = plt.figure(frameon='False')
    # ax = plt.Axes(fig, [0., 0., 1., 1.])
    # ax.set_axis_off()
    # fig.add_axes(ax)
    # ax.imshow(output, aspect='auto')
    # fig.show()
    
```

```python
import os
path = r'D:\scalable_analytics_institute\semantic_depth_transformer\sc_depth_pl\data\ddad\ddad\training'
for number in os.listdir(path):
    if '.' not in number:
        # for all directories in the specified path
        image_folder = os.path.join(path, number)

        # make a new directory to store the new midas depth outputs
        new_depth_folder = os.path.join(image_folder, 'midas_depth')
        if not os.path.isdir(new_depth_folder):
            os.mkdir(new_depth_folder)
        
        for file in os.listdir(image_folder):
            if '.jpg' in file:
                # for all images in the directory
                image_path = os.path.join(image_folder, file)
                save_path = os.path.join(new_depth_folder, file)
                save_depth(image_path, save_path)
                quit()
```
